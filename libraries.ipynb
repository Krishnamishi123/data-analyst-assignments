```python

#Q1. Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the
#dataset information to understand its structure.

import pandas as pd
path = r"C:\Users\91630\Downloads\retail_dataset .csv"

df = pd.read_csv(path)
print('First 5 rows:')
print(df.head())
print('\nDataFrame info:')
print(df.info())
```

```python
#Q2. Inspect the dataset and identify missing values in each column. Display the total number of missing
#values present in the dataset.
print("Missing values per column:")
print(df.isna().sum())          # per-column count

print("\nTotal missing values:")
print(df.isna().sum().sum())    # overall total               
```

```python
#Q3. Replace missing values in the Quantity and Price columns with their respective mean values using Pandas.


# Compute means
quantity_mean = df['Quantity'].mean()
price_mean = df['Price'].mean()

# Fill missing values with mean
df['Quantity'] = df['Quantity'].fillna(quantity_mean)
df['Price']    = df['Price'].fillna(price_mean)

print('Filled Quantity nulls with', quantity_mean)
print('Filled Price nulls with', price_mean)
print('\nMissing values now:')
print(df[['Quantity', 'Price']].isna().sum())
```

```python
#Q4. Remove all rows where the Product Category or Region column contains missing values.


df_clean = df.dropna(subset=['Product Category', 'Region'])


print("Rows before:", len(df))
print("Rows after:", len(df_clean))
print("Missing in Product Category:", df_clean['Product Category'].isna().sum())
print("Missing in Region:", df_clean['Region'].isna().sum())
```

```python
"""
Q5. Create a new column called Revenue using NumPy, where:

Revenue = Quantity × Price

Display the updated dataset with the new column.
"""
import numpy as np
df['Revenue'] = np.multiply(df['Quantity'], df['Price'])
print(df)
```

```python
#Q6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.



total_revenue = np.nansum(df['Revenue'])
print("Total revenue:", total_revenue)
```

```python
#Q7. Group the dataset by Product Category and calculate the total revenue generated by each category.
revenue_by_category=df.groupby('Product Category')['Revenue'].sum()
print(revenue_by_category)
```

```python
#Q8. Based on the revenue calculated for each product category, identify the top 3 categories with the
#highest revenue and the bottom 3 categories with the lowest revenue.
top3=revenue_by_category.sort_values(ascending=False).head(3)
bottom3=revenue_by_category.sort_values(ascending=True).head(3)
print("top3 revenue",top3)
print("bottom3 revenue",bottom3)
```

```python
#Q9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the
#region with the highest revenue and the region with the lowest revenue.
region_revenue = df.groupby('Region', dropna=True)['Revenue'].sum().sort_values(ascending=False)
print(region_revenue)

highest_region = region_revenue.idxmax()
highest_value = region_revenue.max()
lowest_region = region_revenue.idxmin()
lowest_value = region_revenue.min()

print("Highest revenue region:",highest_region ,highest_value)
print("Lowest revenue region:",lowest_region,lowest_value)
```

```python
#Q10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the
#total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column
#using NumPy.
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
df['Month'] = df['Date'].dt.month          
monthly_revenue = df.groupby('Month')['Revenue'].sum()
print("Monthly total revenue:\n", monthly_revenue)
revenue_series = df['Revenue'].dropna()
mean_revenue = np.mean(revenue_series)
median_revenue = np.median(revenue_series)
std_revenue = np.std(revenue_series, ddof=0) 
print("\nRevenue stats:")
print("Mean:  ", mean_revenue)
print("Median:", median_revenue)
print("Std:   ", std_revenue)
```
