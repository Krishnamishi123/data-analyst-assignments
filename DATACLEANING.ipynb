```python
"""
Q1. Missing Data Identification

Scenario:
 The hospital suspects incomplete patient records.

Task:

Identify missing values in each column

Calculate percentage of missing data
"""
import pandas as pd
df=pd.read_csv(r"C:\Users\91630\Downloads\healthcare_data_cleaning_dataset.csv")
missing_values=df.isnull().sum()
total_rows=len(df)
missing_percent=(missing_values/total_rows)*100
print("Missing Values in Each Column:")
print(missing_values)
print("\nPercentage of Missing Data:")
print(missing_percent)
```

```python
"""
Q2. Handling Missing Age

Scenario:
 Age is critical for medical analysis, but some values are missing.

Task:

Replace missing Age values with an appropriate method

Justify your choice (mean/median)
"""
median_age=df['Age'].median()
df['Age'].fillna(median_age, inplace=True)
print("Missing Age values have been replaced with the median age:", median_age)
# i have used median to replace missing age values because age data can often be skewed due to outliers (e.g., very young or very old patients).
#  The median is less affected by outliers compared to the mean, providing a more accurate representation of the central tendency of the age data.
```

```python
"""
Q3. Handling Missing Treatment Cost

Scenario:
 Treatment cost is highly skewed due to expensive treatments.

Task:

Handle missing Treatment_Cost values

Choose the correct imputation method and explain why
"""
median_cost=df['Treatment_Cost'].median()
df['Treatment_Cost'].fillna(median_cost,inplace=True)
print("Missing Treatment_Cost values have been replaced with the median cost:", median_cost)
# I have chosen to replace missing Treatment_Cost values with the median because treatment costs can be highly skewed due to a few expensive treatments. 
# The median is less influenced by extreme values (outliers) than the mean, making it a more appropriate measure of central tendency for skewed data. 
# This approach helps to provide a more accurate representation of the typical treatment cost without being distorted by outliers.
```

```python
"""
Q4. Duplicate Patient Records

Scenario:
 Some patient records were entered multiple times.

Task:

Identify duplicate rows

Remove duplicates

Compare dataset size before and after
"""
initial_size=len(df)
duplicates=df.duplicated()
print("Duplicate Rows:")
print(df[duplicates])
df.drop_duplicates(inplace=True)
final_size=len(df)
print(f"Dataset size before removing duplicates: {initial_size}")
print(f"Dataset size after removing duplicates: {final_size}")
```

```python
"""
Q5. Invalid Age Values (Data Quality Check)

Scenario:
 Some patients have unrealistic age values (e.g., >100 or <0).

Task:

Detect such records

Decide whether to remove or correct them

"""
invalid_age_records=df[(df['Age']>100)|(df['Age']<0)]
print("Records with Invalid Age Values:")
print(invalid_age_records)
# I have decided to remove records with invalid age values because they are likely to be data entry errors and do not represent real patients.
# Removing these records helps to maintain the integrity of the dataset and ensures that analyses based on age are accurate and meaningful.
df.drop(invalid_age_records.index, inplace=True)
```

```python
"""
Q6. Outlier Detection (Treatment Cost)

Scenario:
 Extreme treatment costs are affecting analysis.

Task:

Detect outliers using IQR method

Display number of outliers
"""
Q1=df['Treatment_Cost'].quantile(0.25)
Q3=df['Treatment_Cost'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=df[(df['Treatment_Cost']<lower_bound)|(df['Treatment_Cost']>upper_bound)]
print("Outliers are:",outliers)
print("Number of Outliers in Treatment_Cost:", len(outliers))
```

```python
"""
Q7. Outlier Treatment

Scenario:
The business team wants to retain all records.

Task:

Apply capping (Winsorization) on Treatment_Cost

Use 5th and 95th percentile
"""
lower_cap=df['Treatment_Cost'].quantile(0.05)
upper_cap=df['Treatment_Cost'].quantile(0.95)
df['Treatment_Cost']=df['Treatment_Cost'].apply(lambda x:lower_cap if x<lower_cap else upper_cap if x>upper_cap else x)
print("Treatment cost before applying capping (Winsorization):")
print(df['Treatment_Cost'])
print("Treatment_Cost after applying capping (Winsorization):")
print(df['Treatment_Cost'])

```

```python
"""
Q8. Transformation

Scenario:
 Treatment cost is highly skewed.

Task:

Apply log transformation

Create a new column

Compare before vs after distribution
"""
import numpy as np
df['log_treatment_cost']=np.log(df['Treatment_Cost'])
print("Treatment Cost Distribution Before Log Transformation:")
print(df['Treatment_Cost'].describe())
print("Treatment Cost Distribution After Log Transformation:")
print(df['log_treatment_cost'].describe())
```

```python
"""
Q9. Time-Based Missing Handling

Scenario:
 Admission dates should follow a logical sequence.

Task:

Sort data by Admission_Date

Apply forward fill or backward fill where appropriate

Justify your choice
"""
df['Admission_Date']=pd.to_datetime(df['Admission_Date'])
df.sort_values(by='Admission_Date', inplace=True)
df['Admission_Date'].fillna(method='ffill', inplace=True)
print("Data sorted by Admission_Date and missing values handled with forward fill.")
print(df['Admission_Date'])
# I have chosen to apply forward fill (ffill) for handling missing Admission_Date values because
# it assumes that the missing date is the same as the most recent previous date, which is a reasonable assumption in a time series context.
# This method helps to maintain the chronological order of the data and is appropriate when the missing values
# are likely to be similar to the preceding values, such as in cases where patients are admitted on consecutive days.
```
