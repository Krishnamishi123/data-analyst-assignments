```python
import pandas as pd
df=pd.read_csv(r"C:\Users\91630\Downloads\House Price India.csv")
df.info()
df.describe()
df.isnull().sum()
# No null values int hsi dataset so no requirement of handling the missing values
```

```python
#Q1. What is the average, median, and standard deviation of house prices?
average_price=df['Price'].mean()
median_price=df['Price'].median()
std_dev_price=df['Price'].std()
print(f"Average Price: {average_price}")
print(f"Median Price: {median_price}")
print(f"Standard Deviation of Price: {std_dev_price}")
```

```python
#Q2. Which number of bedrooms is most common?
most_common_bedrooms=df['number of bedrooms'].mode()[0]
print(f"Most Common Number of Bedrooms: {most_common_bedrooms}")
```

```python
#Q3. Check if the price data is skewed.
skewness_price=df['Price'].skew()
print(f"Skewness of Price: {skewness_price}")
if skewness_price > 0:
    print("The price data is positively skewed.")
elif skewness_price < 0:
    print("The price data is negatively skewed.")
else:
    print("The price data is perfectly symmetrical.")
```

```python
#Q4. What is the average price for each number of bedrooms?
average_price_bedrooms=df.groupby('number of bedrooms')['Price'].mean()
print("Average Price for Each Number of Bedrooms:")
print(average_price_bedrooms)
```

```python
#Q5.  What is the relationship between living area and price? (Use : Area without Basement).
correlation_area_price=df['living area'].corr(df['Price'])
print(f"Correlation between Living Area and Price: {correlation_area_price}")
if correlation_area_price > 0:
    print("There is a positive correlation between living area and price.")
elif correlation_area_price < 0:
    print("There is a negative correlation between living area and price.")
else:
    print("There is no correlation between living area and price.")
```

```python
#Q6. Identify any anomalies where houses have high prices but low areas
anomalies=df[(df['Price'] > df['Price'].quantile(0.75)) & (df['Area of the house(excluding basement)'] < df['Area of the house(excluding basement)'].quantile(0.25))]
print("Anomalies (High Price but Low Area):")#here we select top 25% of price and bottom 25% of area to identify anomalies where houses have high prices but low areas
print(anomalies)
```

```python
#Q7. Compare average price based on number of floors and houses with or without waterfront.

average_price_floors=df.groupby('number of floors')['Price'].mean()
print("Average Price Based on Number of Floors:")
print(average_price_floors)
average_price_waterfront=df.groupby('waterfront present')['Price'].mean()
print("Average Price Based on Waterfront:")
print(average_price_waterfront)


```

```python
#Q8. Identify the minimum and maximum house price. What does this indicate?
min_price=df['Price'].min()
max_price=df['Price'].max()
median_price=df['Price'].median()
print(f"Median House Price: {median_price}")
print(f"Minimum House Price: {min_price}")
print(f"Maximum House Price: {max_price}")
#we can see that the minimum price is significantly lower than the median price, and the maximum price is significantly higher than the median price, 
# indicating a wide range of house prices in the dataset. This suggests that there may be some outliers or a skewed distribution of house prices.

```

```python
#Q9. Which location (zipcode/area) has the highest average price?
average_price_location=df.groupby('Postal Code')['Price'].mean()
highest_avg_price_location=average_price_location.idxmax()
print(f"Location with the Highest Average Price: {highest_avg_price_location}")

```
