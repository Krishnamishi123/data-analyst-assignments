```python
#Question 10 : Write a Python program to count how many times a number appears in a list.
#Example List
#[2,3,4,2,5,2]
numbers = [2, 3, 4, 2, 5, 2]
print("The list is:", numbers)
print("enter no from list u want to check count of:")
no=int(input())
count =0
for i in range(len(numbers)):
    if numbers[i]==no:
        count+=1
print(f"The number {no} appears {count} times in the list.")
```
