```python
#Q1. Create a countplot to show the number of passengers by gender

import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv(r"C:\Users\91630\Downloads\Titanic.csv")

df['Age']=df['Age'].fillna(df['Age'].median())
df['Cabin']=df['Cabin'].fillna('Unknown')
#most_freqeunt_embark_town=df['embark_town'].mode()[0]
#df['embark_town']=df['embark_town'].fillna(most_freqeunt_embark_town)
most_freqeunt_embarked=df['Embarked'].mode()[0]
df['Embarked']=df['Embarked'].fillna(most_freqeunt_embarked)

df=sns.load_dataset("titanic")
ax=sns.countplot(x='sex',data=df)
plt.title('Number of Passengers by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
for containers in ax.containers:
    ax.bar_label(containers)
plt.show()


#countplot clearly shows that there were more male passengers than female passengers on the Titanic.
#  The countplot provides a visual representation of the distribution of passengers
```

```python
#Q2. Create a bar chart to show the number of passengers in each passenger class using different colors.
class_counts=df['pclass'].value_counts().sort_index()
bar_colors=['skyblue', 'salmon', 'lightgreen']
plt.figure(figsize=(8,5))
bars=class_counts.plot(kind='bar', color=bar_colors)
for bar in bars.containers:
    bars.bar_label(bar)
plt.title('Number of Passengers by Class')
plt.xlabel('Passenger Class',fontsize=12)
plt.ylabel('Count',fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
#The bar chart shows that the majority of passengers were in the 3rd class, 
# followed by the 1st class, and the least number of passengers were in the 2nd class.
#  The different colors help to visually distinguish between the classes.
```

```python
#Q3. Create a histogram to show the distribution of passenger age.
sns.histplot(df['age'], bins=20, kde=True, color='purple')
plt.xlabel('Age')
plt.title('Distribution of Passenger Age')
plt.show()
skewness=df['age'].skew()
print(f'Skewness of Age Distribution: {skewness}')
#the histogram clearly shows that the age distribution of passengers is right-skewed,
# with a longer tail on the right side.
```

```python
#Q4. Create a boxplot to detect outliers in the fare column.
ax=sns.boxplot(x=df['fare'],color='orange')
plt.title('Box Plot of Fare')
plt.xlabel('Fare')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
#the plot clearly shows that there are several outliers in the fare column, as indicated by the points that lie above the upper whisker of the boxplot.
# These outliers represent passengers who paid significantly higher fares compared to the majority of passengers, which could be due to various factors 
# such as class, cabin type, or special accommodations. The boxplot provides a visual representation of the distribution of fares and helps to identify 
# these outliers effectively.
```

```python
"""
Q5. Create a scatterplot to 
a. show the relationship between age and fare.
b. analyze the relationship between family size (SibSp + Parch) and fare.
"""
# a. Scatterplot for age and fare
plt.figure(figsize=(10,5))
plt.scatter(df['age'],df['fare'],color='blue',alpha=0.5)
plt.title('Scatter Plot of Age vs Fare')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
# The scatterplot shows a positive correlation between age and fare, indicating that older passengers 
#tended to pay higher fares.
# b. Scatterplot for family size and fare
df['family_size']=df['sibsp']+df['parch']
plt.figure(figsize=(10,5))
plt.scatter(df['family_size'], df['fare'], color='green', alpha=0.5)
plt.title('Scatter Plot of Family Size vs Fare')
plt.xlabel('Family Size (SibSp + Parch)')
plt.ylabel('Fare')
plt.grid(True, linestyle='--', alpha=0.7)
plt.show()
# the first scatter plot clearly shows that age range from 20 to 50 pay more 
# The second scatterplot shows a positive correlation between family size and fare,
# indicating that passengers with larger families tended to pay higher fares.
```

```python
#Q6. Create a countplot to show survival count by gender
ax=sns.countplot(x='sex', hue='survived', data=df)
for containers in ax.containers:
    ax.bar_label(containers)
plt.title("Survival Count by gender")
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show() 
#plot clearly shows that a higher proportion of female
#passengers survived compared to male passengers.
```

```python
#Q7. Create a bar chart to show survival by passenger class.
survival_counts_by_class=df.groupby('pclass')['survived'].sum()
ax=survival_counts_by_class.plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])

plt.title('Survival Count by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Count')
plt.xticks(rotation=0)
plt.show()
# The bar chart shows that the survival count was highest for passengers in the 1st class, followed by the 3rd class,
# and lowest for passengers in the 2nd class. This indicates that passengers in higher classes had a better chance of
# survival compared to those in lower classes.

```

```python
#Q8. Create a heatmap to show correlation between numerical variables.
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(numeric_only=True), annot=True,cmap='coolwarm',linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```

```python
#Q9. Identify which passenger class paid the highest fare on average.
average_fare_by_class=df.groupby('pclass')['fare'].mean()
highest_fare_class=average_fare_by_class.idxmax()
print(f"The passenger class that paid the highest average fare is: {highest_fare_class}")
# The passenger class that paid the highest average fare is: 1st class.
```
