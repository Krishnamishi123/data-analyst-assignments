```python
"""
Q1 Dataset Exploration
1)Perform the following exploratory steps:
2)Display the first 10 rows
3)Display the dataset structure using info()
4)Find the number of rows and columns
5)Identify the data types of each column
"""

import pandas as pd
df=pd.read_csv(r"C:\Users\91630\Downloads\Advanced_pandas_sales_dataset.csv")
# Display the first 10 rows
print(df.head(10))
# Display the dataset structure using info()
print(df.info())
# Find the number of rows and columns
print("Number of rows:", df.shape[0])
print("Number of columns:", df.shape[1])
# Identify the data types of each column
print(df.dtypes)
```

```python
"""
Q2 Sales Filtering Analysis
Filter the dataset to display:
1)Orders where Sales > 3500
2)Orders where Region = "East" and Sales > 2500
3)Explain what insights you observe.
"""
sale_greater_than_3500=df[df["Sales"] > 3500]
print(sale_greater_than_3500)
east_sales_greater_than_3500=df[(df["Region"]=="East")&(df["Sales"]>3500)]
print(east_sales_greater_than_3500)
# Insights:
# 1) Orders with Sales > 3500 indicate high-value transactions, which may be crucial 
# for revenue analysis and identifying key customers.
# 2) Orders in the East region with Sales > 3500 suggest that this region may have a strong market presence or high 
# demand for certain products, which could inform regional marketing strategies and inventory management.
```

```python
"""
Q3 Customer Purchase Analysis
Identify the top 5 customers based on total sales.
Hint: Use GroupBy with aggregation.
"""
top_customers=df.groupby("Customer")["Sales"].sum().sort_values(ascending=False).head(5)
print(top_customers)
```

```python
"""
Q4 Regional Sales Performance
Calculate the total sales generated by each region.
Then identify:
Which region generated the highest revenue?
"""
regional_sales=df.groupby("Region")["Sales"].sum()
print(regional_sales)
highest_revenue_region=regional_sales.idxmax()
print("Region with highest revenue:", highest_revenue_region)
```

```python
#Q5 Product Performance Analysis
#Calculate the total quantity sold for each product.
#Which product appears to be the most popular?
product_quantity_sold=df.groupby("Product")["Quantity"].sum()
print(product_quantity_sold)
most_popular_prod=product_quantity_sold.idxmax()
print("Most popular product:", most_popular_prod)
```

```python
#Q6 Multi-Level Grouping
#Group the dataset by: Region and Category
#Calculate the total sales for each combination.
#What insights can you derive from this analysis?
multilevel_group=df.groupby(["Region","Category"])["Sales"].sum()
print(multilevel_group)
best_combination=multilevel_group.idxmax()
print("Best combination:", best_combination)
# Insights:
# 1) Grouping by Region and Category allows us to identify which combinations of region and category are driving the 
# most sales, which can inform targeted marketing and inventory decisions.
# this reveals that east and electronics combination performs the best, suggesting a strong demand for electronics in 
# the East region, which could lead to focused promotional efforts and stock management in that area.
```

```python
"""
Q7 Pivot Table Creation
Create a pivot table showing:
Rows → Region
Columns → Product
Values → Total Sales
Which region sells the most laptops?
"""
pivot_table=df.pivot_table(index="Region",columns="Product",values="Sales",aggfunc="sum")
print(pivot_table)
laptop_sales_most_regionly=pivot_table["Laptop"].idxmax()
print("Region that sells the most laptops:", laptop_sales_most_regionly)
```

```python
#Q8 Profitability Analysis
#Calculate the average profit for each product category.
#Which category is most profitable?
avg_prfit_category=df.groupby("Category")["Profit"].mean()
print(avg_prfit_category)
most_profitable_category=avg_prfit_category.idxmax()
print("Most profitable category:", most_profitable_category)
```

```python
"""
Q9 Discount Impact Study
Analyze the relationship between discount and profit.
Tasks:
Calculate the average profit for each discount level
Identify whether higher discounts lead to lower profit
"""
avg_profit_discount = df.groupby("Discount")["Profit"].mean()
print(avg_profit_discount)
corr = df["Discount"].corr(df["Profit"])
print("Discount vs Profit correlation:", corr)
# the correlation coefficient is negative strongly that shows that higher discounts are associated with lower profits, 
# indicating a potential trade-off between offering discounts and maintaining profitability.
```

```python
"""
Q10 End-to-End Data Wrangling Task
Perform the following workflow:
Filter the dataset where Sales > 3000
Group data by Region and Product
Calculate total sales
Reshape the result into a pivot table
Explain the business insights obtained from this analysis.
"""
filter_data_sales_gt_3000=df[df["Sales"]>3000]
grouped_data=filter_data_sales_gt_3000.groupby(["Region", "Product"])["Sales"].sum()
pivot_table=grouped_data.unstack(level="Product")
print(pivot_table)

# Insights:
# 1) This analysis helps identify which regions and products are contributing the most to high sales, allowing 
# for targeted marketing and inventory strategies.
# 2) It can reveal regional preferences for certain products, guiding product placement and promotional efforts
# 3) we understand from data that foe each region which category gives highest sales
#as can be seen that for east it is laptop,for north region it is phone,for souht it is mouse and for west it is
#keyboard
```
