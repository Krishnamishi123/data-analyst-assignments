```python
"""
Q1. What is Machine Learning? Explain the main types of Machine Learning with examples.
Machine Learning is a subset of artificial intelligence that enables computers to learn from data and make decisions or predictions without 
being explicitly programmed. It involves the development of algorithms that can identify patterns in data and improve their performance 
over time as they are exposed to more data.
The main types of Machine Learning are:
1. Supervised Learning: In supervised learning, the algorithm is trained on a labeled dataset, where each input is paired with the correct 
output. The goal is to learn a mapping from inputs to outputs. Examples include:
   - Classification: Predicting whether an email is spam or not based on its content.
    - Regression: Predicting the price of a house based on its features such as size, location, and number of bedrooms.
2. Unsupervised Learning: In unsupervised learning, the algorithm is trained on an unlabeled dataset, where the goal is to find hidden 
patterns or structures in the data. Examples include:
   - Clustering: Grouping customers based on their purchasing behavior to identify distinct segments.
    - Dimensionality Reduction: Reducing the number of features in a dataset while retaining important information, such as using Principal 
    Component Analysis (PCA).
3. Reinforcement Learning: In reinforcement learning, an agent learns to make decisions by interacting with an environment and receiving 
feedback in the form of rewards or penalties. The goal is to learn a policy that maximizes cumulative rewards over time. Examples include:
   - Game Playing: Teaching an AI to play chess or Go by rewarding it for winning and penalizing it for losing.
    - Robotics: Training a robot to navigate a maze by rewarding it for reaching the exit and penalizing it for collisions.

#Q2. What is Supervised Learning? Explain the difference between Regression and Classification.
Supervised Learning is a type of machine learning where the algorithm is trained on a labeled dataset, meaning that each input is paired 
with the correct output. The goal of supervised learning is to learn a mapping from inputs to outputs so that the model can make accurate 
predictions on new, unseen data.
The main difference between Regression and Classification in supervised learning is the type of output they predict:
- Regression: In regression, the output is a continuous value. The model predicts a numerical value based on the input features. For 
example, predicting the price of a house based on its size, location, and number of bedrooms is a regression problem.
- Classification: In classification, the output is a discrete class label. The model predicts which category or class an input belongs to. 
For example, predicting whether an email is spam or not based on its content is a classification problem.

#Q3. What is Linear Regression? Write the Linear Regression equation and explain intercept and coefficient.
Linear Regression is a statistical method used to model the relationship between a dependent variable and one or more independent variables.
It assumes that there is a linear relationship between the input features and the output variable. The goal of linear regression is to find
the best-fitting line that minimizes the sum of squared differences between the observed values.
The Linear Regression equation is:
y = β0 + β1x1 + β2x2 + ... + βnxn + ε
Where:
- y is the dependent variable (the value we want to predict)
- β0 is the intercept (the value of y when all independent variables are 0)
- β1, β2, ..., βn are the coefficients (the change in y for a one-unit change in each independent variable)
- x1, x2, ..., xn are the independent variables (the features used to make predictions)
- ε is the error term (the difference between the observed and predicted values)

#Q4. Explain the following evaluation metrics used in Linear Regression:
# SSE (Sum of Squared Errors)
# MSE (Mean Squared Error)
# RMSE (Root Mean Squared Error)
# R² Score (Coefficient of Determination)
# Adjusted R²
SSE (Sum of Squared Errors): SSE is the sum of the squared differences between the observed values and the predicted values. It measures the 
total error in the model. A lower SSE indicates a better fit of the model to the data.
MSE (Mean Squared Error): MSE is the average of the squared differences between the observed values and the predicted values. It provides a 
measure of the average error in the model. A lower MSE indicates a better fit of the model to the data.
RMSE (Root Mean Squared Error): RMSE is the square root of the MSE. It provides a measure of the average error in the same units as the
dependent variable. A lower RMSE indicates a better fit of the model to the data.
R² Score (Coefficient of Determination): R² is a statistical measure that represents the proportion of the variance in the dependent variable
that is explained by the independent variables in the model. It ranges from 0 to 1, where a higher R² indicates a better fit of the 
model to the data.
Adjusted R²: Adjusted R² is a modified version of R² that takes into account the number of independent variables in the model. It adjusts for
the number of predictors and provides a more accurate measure of the model's performance, especially when comparing models with different 
numbers of predictors. A higher Adjusted R² indicates a better fit of the model to the data, while penalizing for adding unnecessary predictors.

#Q5. Explain the Machine Learning workflow steps from data collection to model evaluation.
The Machine Learning workflow typically consists of the following steps:
1.Data Collection: The first step is to gather relevant data for the problem at hand. This can involve collecting data from 
various sources such as databases, APIs, or web scraping.
2.Data Preprocessing: Once the data is collected, it needs to be cleaned and prepared for analysis. This step may involve handling missing 
values, encoding categorical variables, normalizing or standardizing features, and splitting the data into training and testing sets.
3.Feature Engineering: In this step, new features may be created from the existing data to improve the performance of the model. This can 
involve techniques such as polynomial features, interaction terms, or dimensionality reduction.
4.Model Selection: The next step is to choose an appropriate machine learning algorithm based on the problem type (e.g., regression, 
classification) and the characteristics of the data. Common algorithms include linear regression, decision trees, support vector machines, 
and neural networks.
5.Model Training: The selected model is trained on the training dataset, where it learns the patterns and relationships in the data.
 This involves optimizing the model's parameters to minimize the error between the predicted and actual values.
6.Model Evaluation: After training, the model's performance is evaluated using the testing dataset. This step involves calculating 
evaluation metrics such as accuracy, precision, recall, F1 score for classification problems, or MSE, RMSE, R² for regression problems. 
The evaluation results help determine how well the model generalizes to new data and whether it is suitable for deployment.
7.Model Tuning: If the model's performance is not satisfactory, hyperparameter tuning can be performed to optimize the model's parameters
and improve its performance. This can involve techniques such as grid search or random search.
8.Model Deployment: Once the model is trained and evaluated, it can be deployed to make predictions on new, unseen data. This can involve
integrating the model into a production environment, such as a web application or an API, to provide real-time predictions to users.
9.Monitoring and Maintenance: After deployment, it is important to continuously monitor the model's performance and update it as needed. This
can involve retraining the model with new data, addressing any issues that arise, and ensuring that the model continues to perform well over time.
"""


```

```python
"""
Q6. Load the Boston Housing dataset and perform basic data exploration using:
head()
info()
describe()
"""
import pandas as pd
df=pd.read_csv(r"C:\Users\91630\Downloads\BostonHousing.csv")
df.head()
df.info()
df.describe()
```

```python
"""
Q7. Perform Exploratory Data Analysis:
Create pairplot
Create correlation heatmap
Identify which features are highly correlated with house price (medv)
"""
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()


price_correlations = correlation_matrix['medv'].sort_values(ascending=False)
print("Correlation with MEDV:\n", price_correlations)


important_features = ['medv', 'rm', 'lstat', 'ptratio']
sns.pairplot(df[important_features])
plt.show()
#rm,zn,b,dis,chas are highly correlated with medv(house price).  
      

```

```python
#Q8. Select independent variables (features) and dependent variable (target).Target variable medv(house price).
x=df.drop('medv',axis=1)
y=df['medv']

print("Independent Variables (Features):\n", x.head())
print("\nDependent Variable (Target):\n", y.head())
```

```python
#Q9. Split the dataset into training and testing sets (80% training, 20% testing).
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print("Training set size:", x_train.shape[0])
print("Testing set size:", x_test.shape[0])
print("Training set size:", y_train.shape[0])
print("Testing set size:", y_test.shape[0])
```

```python
"""
Q10. Train a Linear Regression model and perform the following:

Make predictions

Calculate MAE

Calculate MSE

Calculate RMSE

Calculate R² Score

Calculate Adjusted R²

Write the Linear Regression equation

Interpret the model coefficients 
"""
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np
model=LinearRegression()
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)
r2=r2_score(y_test,y_pred)
n=len(x_test)
p=len(x_test.columns)
adjusted_r2=1-(1-r2)*(n-1)/(n-p-1)
print("Mean Absolute Error (MAE):", mae)
print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("R² Score:", r2)
print("Adjusted R²:", adjusted_r2)
intercept=model.intercept_
coefficients=model.coef_
print("Intercept:", intercept)
print("Coefficients:", coefficients)
```
