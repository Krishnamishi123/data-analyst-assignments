```python
"""
Q1. Creating a DataFrame
Create a Pandas DataFrame using a Python dictionary with the following columns:
Product
Region
Sales
Quantity
Add at least 5 rows of sample data and display the DataFrame.
"""
import pandas as pd
data = {
    "Product": ["table", "chair", "bucket", "Ceiling fan", "Lamp"],
    "Region": ["North", "South", "East", "West", "Central"],
    "Sales": [100, 200, 150, 300, 250],
    "Quantity": [10, 20, 15, 30, 25]
}
df = pd.DataFrame(data)
print(df)
```

```python
"""
Q7. Filtering Data
Filter the dataset to display:
1.Rows where Region = "East".
"""
east=df1[df1["Region"]=="East"]
print(east)
```

```python
#Rows where Sales > 1500.
sales_1500=df1[df1["Sales"]>1500]
print(sales_1500)
```

```python
#Q8. Multiple Condition Filtering
#Display rows where:
#Region is West and
#Sales is greater than 1200
west_sales_1200=df1[(df1["Region"]=="West")&(df1["Sales"]>1200)]
print(west_sales_1200)
```

```python
"""
Q9. Descriptive Statistics
Generate summary statistics using:
describe()
From the result identify:
Mean sales value
Maximum quantity sold
Minimum sales value
"""
df1.describe()
mean_sales_value=df1["Sales"].mean()
max_quantity_sold=df1["Quantity"].max()
min_sales_value=df1["Sales"].min()
print("Mean sales value:", mean_sales_value)
print("Maximum quantity sold:", max_quantity_sold)
print("Minimum sales value:", min_sales_value)
```

```python
"""
Q10. Aggregation Analysis
Using the Sales column, calculate:
Total sales
Average sales
Maximum sales
Minimum sales
"""
tota_sales=df1["Sales"].sum()
average_sales=df1["Sales"].mean()
maximum_sale=df1["Sales"].max()
minimum_sale=df1["Sales"].min()
print("Total sales:", tota_sales)
print("Average sales:", average_sales)
print("Maximum sales:", maximum_sale)
print("Minimum sales:", minimum_sale)
```
